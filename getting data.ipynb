```python
import fantasy as fp
# noinspection PyUnresolvedReferences
from fantasyalgo import best_subset
import datetime as dt
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
start = dt.date(2022, 1, 1)
end = dt.date.today()
delta = dt.timedelta(days=90)

events = [6586, 6140, 6136, 6621, 6370]
cur = 0

dataframes = []
used = []

for key in events:

    event = fp.Event(key)
    players = event.get_players()

    for p in players:
        if p.key not in used:
            used.append(p.key)
            dataframes.append(p.get_dataset(start, end, delta))

    cur += 1
```

```python
data = pd.concat(dataframes)
data = data.loc[~data['event'].str.contains("ESL Pro League")]
data.shape
```

```python
set_first = data.drop(columns=['player', 'event', 'pts']).to_numpy()
pts = data['pts'].to_numpy()
```

```python
event = fp.Event(6970)
players = event.get_players()
len(players)
```

```python
costs_data = event.get_costs("costs_data.html")
len(costs_data)
```

```python
for p in players:
    if p.name not in list(costs_data.keys()):
        name = f"{p.name}-"
        if name in list(costs_data.keys()):
            p.name = name
            print(p.name)
```

```python
costs = []
keys = list(costs_data.keys())
not_found = []

cnt = 0
for p in players:
    if p.name in keys:
        costs.append(costs_data[p.name])
    else:
        not_found.append(cnt)
    cnt += 1
len(costs)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor

dtree = DecisionTreeRegressor(max_depth=1)
linear = LinearRegression()
dtree.fit(set_first, pts)
linear.fit(set_first, pts)
```

```python
predicted_linear = []
rank = event.rank

for p in players:
    all_rating = p.get_stats(dt.date.today() - dt.timedelta(days=90), dt.date.today(), fp.EventFilter.ALL)
    big_rating = p.get_stats(dt.date.today() - dt.timedelta(days=90), dt.date.today(), fp.EventFilter.BIG)
    vector = np.array([[rank, all_rating, big_rating]])
    predicted_linear.append(float(linear.predict(vector)[0]))
print(len(predicted_linear))
```

```python
cnt = 0
names = list(costs_data.keys())


for p in players:
    if p.name not in names:
        print(cnt)
    cnt += 1
```

```python
linear_team = best_subset(predicted_linear, costs)
```

```python
for i in linear_team:
    print(f"{players[i].name} --> {predicted_linear[i]}")
```
