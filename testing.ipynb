```python
import fantasyparse as fp
# noinspection PyUnresolvedReferences
from fantasyalgo import best_subset
import datetime as dt
import numpy as np
import pandas as pd
```

```python
# Parameters
event = fp.Event(6713)
```

```python
end = event.start_date
start = end - dt.timedelta(days=90)
efilter = fp.EventFilter.ALL
rfilter = fp.RankFilter.ALL
delta = dt.timedelta(days=90)
```

```python
costs = event.get_costs("RMRA.html")
len(costs)
```

```python
players = event.get_players()
len(players)
```

```python
for p in players:
    if p.name not in list(costs.keys()):
        name = f"{p.name}-"
        if name in list(costs.keys()):
            p.name = name
            print(p.name)
```

```python
data = event.process(players, costs, start, end, efilter, rfilter, delta)
len(data)
```

```python
data
```

```python
data
```

```python
predictions, costs = [], []
for pair in data.values():
    predictions.append(pair.pts)
    costs.append(pair.cost)
print(len(predictions), len(costs))
```

```python
result = best_subset(predictions, costs)
result
```

```python
names = list(data.keys())
answer = 0
for i in result:
    answer += data[names[i]].pts
    print(names[i])
answer
```

```python
answer_points = []
names = []
for p in players:
    names.append(p.name)
    answer_points.append(p.calc_pts(event.key, fp.RankFilter.ALL))
print(len(names), len(answer_points))
```

```python
answer_subset = best_subset(answer_points, costs)
answer = 0
for i in answer_subset:
    answer += answer_points[i]
    print(names[i])
answer
```
